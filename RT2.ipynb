**Empezamos Fourier**

```python
# Creamos una copia del df limpio
df_1F = df_1.copy()

# Cambiamos las columnas categoricas de Brand y Geo a string 
df_1F["Brand"] = df_1F["Brand"].astype(str).str.strip()
df_1F["Geo"]   = df_1F["Geo"].astype(str).str.strip()

# Crear columna Entidad = Brand_Geo
df_1F["Entidad"] = (df_1F["Brand"] + "_" + df_1F["Geo"]).str.replace(r"\s+", "", regex=True)


# definimos qué columnas son aditivas vs promedios
sum_cols = ["Sales_Value","TV_Impressions","YouTube_Impressions","Facebook_Impressions",
            "Instagram_Impressions","Print_Readership","Radio_Listenership","Trade_Spend"]

avg_cols = ["Net_Price","Weighted_Distribution","CPI","GDP_Growth","Festival_Index","Rainfall_Index"]

flag_cols = [c for c in ["Feature_Flag","Display_Flag","TPR_Flag"] if c in df_1F.columns]

# acompamos por Entidad y semana
df_we = (
    df_1F
    .groupby(["Week","Entidad"], as_index=False)
    .agg({
        **{c: "sum"  for c in sum_cols},   # aditivas
        **{c: "mean" for c in avg_cols},   # continuas (precio, dist, índices)
        **{c: "mean" for c in flag_cols},  # flags → proporción (intensidad)
    })
    .rename(columns={c: c.replace("_Flag","_Rate") for c in flag_cols})
    .sort_values(["Entidad","Week"])
)

# revisamos el df transformado
df_we

```

```python
# Partimos de df_we ya existente
df_1F2 = df_we.copy()

rate_cols = [c for c in ["Feature_Rate","Display_Rate","TPR_Rate"] if c in df_1F2.columns]

# calendario semanal continuo por Entidad 
def weekly_reindex_fill(g):
    g = g.sort_values("Week").set_index("Week").asfreq("W-MON")
    if sum_cols:  g[sum_cols] = g[sum_cols].fillna(0)
    for cols in (avg_cols, rate_cols):
        if cols:   g[cols] = g[cols].interpolate(limit_direction="both")
    return g.reset_index()

# aplicamos la función a cada Entidad y reordenamos
dfc = (df_1F2.groupby("Entidad", group_keys=False).apply(weekly_reindex_fill).sort_values(["Entidad","Week"]).reset_index(drop=True))

# declaramos parámetros para las series de Fourier
P = 52
K =  3

# agregamos las series de Fourier
def add_fourier(g):
    t = np.arange(len(g), dtype=float)
    for k in range(1, K+1):
        g[f"fourier_sin_{k}"] = np.sin(2*np.pi*k*t/P)
        g[f"fourier_cos_{k}"] = np.cos(2*np.pi*k*t/P)
    return g

# aplicamos la función a cada Entidad
dfc = dfc.groupby("Entidad", group_keys=False).apply(add_fourier)

# visualizamos el resultado
dfc
```

**Calculamos adstock**

```python

ENTIDAD = "BRANDA_NORTH"    
TARGET  = "Sales_Value"
MEDIA_COLS = ["TV_Impressions","YouTube_Impressions","Facebook_Impressions","Instagram_Impressions","Print_Readership","Radio_Listenership","Trade_Spend"]


ALPHA = 0.6   
L     = 8 

dfe = (df_we.loc[df_we["Entidad"] == ENTIDAD].sort_values("Week").reset_index(drop=True))

def adstock_geometric(x, alpha=0.6, L=8):
    x = np.asarray(x, float)
    w = alpha ** np.arange(L)              
    xp = np.pad(x, (L-1, 0))               
    out = np.empty_like(x, dtype=float)
    for t in range(len(x)):
        ventana = xp[t:t+L][::-1]          
        out[t] = ventana @ w               
    return out

dfe_ad = dfe[["Week","Entidad", TARGET]].copy()

for col in MEDIA_COLS:
    dfe_ad[f"{col}__ad"] = adstock_geometric(dfe[col].to_numpy(), alpha=ALPHA, L=L)

# ver resultado
dfe_ad
```

```python
# Parámetros de carryover
ALPHA = 0.6   # persistencia (0–1); 0.6 es razonable
L     = 8     # ventana de memoria 

# Canales de media a transformar 
MEDIA_COLS = ["TV_Impressions","YouTube_Impressions","Facebook_Impressions","Instagram_Impressions","Print_Readership","Radio_Listenership","Trade_Spend"]
TARGET = "Sales_Value"

# Función de adstock geométrico 
def adstock_geometric(x, alpha=0.6, L=8):
    x = np.asarray(x, float)
    w = alpha ** np.arange(L)            
    xp = np.pad(x, (L-1, 0))             
    out = np.empty_like(x, dtype=float)
    for t in range(len(x)):
        ventana = xp[t:t+L][::-1]        
        out[t] = ventana @ w             
    return out

# Asegura frecuencia semanal continua por entidad 
def pad_weekly(g):
    g = g.sort_values("Week").set_index("Week").asfreq("W-MON")
    
    # Sin medios = 0 en semanas faltantes
    for c in MEDIA_COLS:
        if c in g: g[c] = g[c].fillna(0)
    
    # Ventas: interpola para mantener continuidad (solo para el cálculo, no "inventes" ventas finales)
    if TARGET in g:
        g[TARGET] = g[TARGET].interpolate().ffill().bfill()
    return g.reset_index()

df_we["Week"] = pd.to_datetime(df_we["Week"], errors="coerce")
df_we2 = (df_we.groupby("Entidad", group_keys=False).apply(pad_weekly).sort_values(["Entidad","Week"]))

# Calcula adstock por entidad para cada canal 
df_ad = df_we2[["Week","Entidad", TARGET] + MEDIA_COLS].copy()
for col in MEDIA_COLS:
    df_ad[f"{col}__ad"] = (df_ad.groupby("Entidad", group_keys=False)[col].transform(lambda s: adstock_geometric(s.to_numpy(), ALPHA, L)))

df_ad
```

```python
# Configuracion basica
assert "Entidad" in df_ad.columns, "df_ad debe tener columna 'Entidad'."
assert "Sales_Value" in df_ad.columns, "df_ad debe tener columna 'Sales_Value'."

media_ad = [c for c in df_ad.columns if c.endswith("__ad")]
if not media_ad:
    raise ValueError("No encuentro columnas con sufijo '__ad' en df_ad.")

# Nos aseguramos de que sean numéricas
for c in ["Sales_Value"] + media_ad:
    df_ad[c] = pd.to_numeric(df_ad[c], errors="coerce")

# CORR POR ENTIDAD Y CANAL (robusto)
rows = []
for entidad, g in df_ad.groupby("Entidad"):
    y = g["Sales_Value"].to_numpy()
    y_std = np.nanstd(y)
    for ch in media_ad:
        x = g[ch].to_numpy()
        # si alguna serie es constante, corr = NaN
        if len(x) < 2 or np.nanstd(x) == 0 or y_std == 0:
            corr = np.nan
        else:
            # np.corrcoef ignora NaNs si los limpias:
            mask = ~np.isnan(x) & ~np.isnan(y)
            if mask.sum() < 2:
                corr = np.nan
            else:
                corr = np.corrcoef(x[mask], y[mask])[0, 1]
        rows.append({"Entidad": entidad, "canal": ch, "corr": corr})

corr_df = pd.DataFrame(rows).dropna(subset=["corr"])

corr_df
```

```python
def plot_corr_heatmap(corr_df, order_channels="mean_abs"):
    # Tabla ancho: filas=Entidad, cols=canal
    pv = corr_df.pivot(index="Entidad", columns="canal", values="corr")

    # Orden de canales (columnas)
    if order_channels == "mean_abs":
        col_order = pv.abs().mean(0).sort_values(ascending=False).index  # canales + relevantes en promedio
    elif order_channels == "mean":
        col_order = pv.mean(0).sort_values(ascending=False).index
    else:
        col_order = pv.columns

    # Orden de entidades (filas): por su correlación máxima en absoluto
    row_order = pv.abs().max(1).sort_values(ascending=False).index

    pv = pv.loc[row_order, col_order]

    # Tamaño acorde a la cantidad
    plt.figure(figsize=(1.2*len(col_order)+2, 0.35*len(row_order)+2))
    sns.heatmap(pv, vmin=-1, vmax=1, center=0, cmap="RdBu_r",
                linewidths=0.3, linecolor="white")
    plt.title("Correlación (Pearson) por entidad y canal (adstock)")
    plt.xlabel("Canal")
    plt.ylabel("Entidad")
    plt.tight_layout()
    plt.show()

# Ejemplo:
plot_corr_heatmap(corr_df)
```

```python
def plot_corr_bars_entity(corr_df, entidad, top=None):
    sub = corr_df[corr_df["Entidad"] == entidad].copy()
    if sub.empty:
        raise ValueError(f"Entidad '{entidad}' no encontrada.")

    # Si quieres quedarte con los top por magnitud:
    if top is not None:
        sub = sub.loc[sub["corr"].abs().nlargest(top).index]

    sub = sub.sort_values("corr")  # de negativo a positivo

    plt.figure(figsize=(8, 0.4*len(sub) + 2))
    plt.barh(sub["canal"], sub["corr"])
    plt.axvline(0, lw=1, color="k")
    plt.title(f"Correlaciones por canal — {entidad}")
    plt.xlabel("Correlación con Sales_Value")
    plt.ylabel("Canal")
    plt.tight_layout()
    plt.show()

# Ejemplo:
plot_corr_bars_entity(corr_df, "BRANDA_CENTRAL", top=8)
plot_corr_bars_entity(corr_df, "BRANDC_NORTHEAST", top=8)
```

```python
def plot_corr_dots_by_channel(corr_df, canal):
    sub = corr_df[corr_df["canal"] == canal].copy()
    if sub.empty:
        raise ValueError(f"Canal '{canal}' no encontrado.")

    sub = sub.sort_values("corr")
    y = np.arange(len(sub))

    plt.figure(figsize=(9, 0.35*len(sub) + 2))
    plt.scatter(sub["corr"], y)
    plt.axvline(0, lw=1, color="k")
    plt.yticks(y, sub["Entidad"])
    plt.title(f"Correlación por entidad — {canal}")
    plt.xlabel("Correlación con Sales_Value")
    plt.ylabel("Entidad")
    plt.tight_layout()
    plt.show()

# Ejemplo:
plot_corr_dots_by_channel(corr_df, "TV_Impressions__ad")
plot_corr_dots_by_channel(corr_df, "Instagram_Impressions__ad")
```

```python
# Elige las columnas de adstock 

ad_cols = ['TV_Impressions__ad', 'YouTube_Impressions__ad', 'Facebook_Impressions__ad', 'Instagram_Impressions__ad', 'Print_Readership__ad', 'Radio_Listenership__ad', 'Trade_Spend__ad']

# Parámetros de la curva Hill
ALPHA = 1.5      
P_K   = 0.50    

def hill(x, alpha, k):
    x = np.clip(x.astype(float), 0, None)        
    if k <= 0:                                   
        return np.zeros_like(x, dtype=float)
    xa = np.power(x, alpha)
    ka = k**alpha
    return xa / (xa + ka)

# Calcula k por canal y aplica Hill (crea columnas __sat)

params = []  

for c in ad_cols:
    
    pos = df_ad.loc[df_ad[c] > 0, c]
    if len(pos) == 0:
        k = df_ad[c].max()
    else:
        k = float(pos.quantile(P_K))
        if not np.isfinite(k) or k <= 0:
            k = float(pos.mean()) if pos.mean() > 0 else float(df_ad[c].max())

    
    if c.endswith("__ad"):
        out = c[:-4] + "__sat"
    elif c.endswith("_ad"):
        out = c[:-3] + "__sat"
    else:
        out = c + "__sat"

    df_ad[out] = hill(df_ad[c].values, ALPHA, k)
    params.append({"canal": c, "alpha": ALPHA, "k": k})


sat_params = pd.DataFrame(params).sort_values("canal")
sat_params

# (opcional) guarda a disco
# df_ad.to_csv("df_con_adstock_y_saturacion.csv", index=False)
```

```python

# Mapea canal -> (alpha, k)
params = dict(zip(sat_params["canal"], zip(sat_params["alpha"], sat_params["k"])))

def hill(x, alpha, k):
    x = np.clip(np.asarray(x, dtype=float), 0, None)
    if not (np.isfinite(k) and k > 0):
        return np.zeros_like(x)
    if not (np.isfinite(alpha) and alpha > 0):
        alpha = 1.0
    xa = np.power(x, alpha)
    return xa / (xa + (k ** alpha))

# crea columnas saturadas por canal
ad_cols = [c for c in df_ad.columns if c.endswith("__ad")]

for c in ad_cols:
    if c not in params:
        print(f" No tengo parámetros (alpha,k) para {c}; lo salto.")
        continue
    a, k = params[c]
    df_ad[c.replace("__ad", "__sat")] = hill(df_ad[c], a, k)

# chequeo rápido
df_ad[[c for c in df_ad.columns if c.endswith("__sat")]].describe()

df_ad.head()   


```

```python
from sklearn.linear_model import LinearRegression


df_final = df_ad.copy().sort_values(["Entidad", "Week"]).reset_index(drop=True)

target = "Sales_Value"

media_cols =[]
for c in df_final.columns:
    if c.endswith("__sat"):
        media_cols.append(c)
        
controls =["Weighted_Distribution","CPI","GDP_Growth","Festival_Index","Rainfall_Index","Feature_Rate","Display_Rate","TPR_Rate",
            "fourier_sin_1","fourier_cos_1","fourier_sin_2","fourier_cos_2","fourier_sin_3","fourier_cos_3"]

features = media_cols + controls

coefs_list = []
preds_list = []

for ent, g in df_final.groupby("Entidad", sort=False):
    X = g[features].astype(float).values   
    y = g[target].astype(float).values

    model = LinearRegression()
    model.fit(X, y)
    yhat = model.predict(X)
    r2 = model.score(X, y)

    coefs_list.append(pd.DataFrame({
        "Entidad": ent,
        "feature": ["(intercept)"] + features,
        "coef": np.r_[model.intercept_, model.coef_],
        "R2_in_sample": r2
    }))

    preds_list.append(pd.DataFrame({
        "Entidad": ent,
        "Week": g["Week"].values,
        "y_true": y,
        "y_hat": yhat,
        "R2_in_sample": r2
    }))

coefs_df  = pd.concat(coefs_list, ignore_index=True)
fitted_df = pd.concat(preds_list, ignore_index=True)

print(coefs_df.groupby("Entidad")["R2_in_sample"].first().sort_values(ascending=False).head())

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score



def fit_and_predict_group(g):
    X = g[features].values    # features = medias __sat + controles + fourier
    y = g[target].values      # target = "Sales_Value"
    m = LinearRegression().fit(X, y)
    g = g.copy()
    g["y_hat"] = m.predict(X)     # predicción de ventas por semana
    g["resid"] = y - g["y_hat"]   # residuo
    g["R2_in_sample"] = r2_score(y, g["y_hat"])
    return g

preds = (df_final.groupby("Entidad", group_keys=False).apply(fit_and_predict_group))

entidad_objetivo = "BRANDA_NORTH"   

# Ejemplo: ver algunas filas
#preds.loc[preds["Entidad"].eq(preds["Entidad"].iloc[156]), ["Week","Entidad","Sales_Value","y_hat","resid","R2_in_sample"]].head(100)

preds.loc[preds["Entidad"] == entidad_objetivo, ["Week", "Entidad", "Sales_Value", "y_hat", "resid", "R2_in_sample"]].head(100)

```

```python
df = df_final.copy().sort_values(["Entidad","Week"])

target = "Sales_Value"
media_cols = [c for c in df.columns if c.endswith("__sat")]

controls = [
    "Weighted_Distribution","CPI","GDP_Growth","Festival_Index","Rainfall_Index",
    "Feature_Rate","Display_Rate","TPR_Rate",
    "fourier_sin_1","fourier_cos_1","fourier_sin_2","fourier_cos_2","fourier_sin_3","fourier_cos_3"
]
controls = [c for c in controls if c in df.columns]  # por si acaso
features = media_cols + controls

# ---- 1) Fit por Entidad + contribuciones semanales
partes = []
coef_table = []

for ent, g in df.groupby("Entidad"):
    X = g[features].values
    y = g[target].values
    m = LinearRegression().fit(X, y)

    coefs = pd.Series(m.coef_, index=features)
    y_hat = m.predict(X)

    # Contribución semanal por canal (solo medios)
    contrib = g[media_cols].mul(coefs[media_cols], axis=1)
    contrib.columns = [c.replace("__sat","__contr") for c in contrib.columns]

    gg = g.copy()
    gg["y_hat"] = y_hat
    gg = pd.concat([gg, contrib], axis=1)
    partes.append(gg)

    # guardamos coeficientes por entidad (útil para interpretación)
    coef_table.append(pd.DataFrame({"Entidad": ent, "feature": coefs.index, "beta": coefs.values}))

df_with_contr = pd.concat(partes, ignore_index=True)
betas_por_entidad = pd.concat(coef_table, ignore_index=True)

# ---- 2) Resumen por entidad
contr_cols = [c for c in df_with_contr.columns if c.endswith("__contr")]
resumen_entidad = (
    df_with_contr.groupby("Entidad")[contr_cols + ["y_hat"]]
    .sum()
    .rename(columns={"y_hat":"y_hat_total"})
)

# ---- 3) Resumen global (todas las entidades)
resumen_global = pd.DataFrame({
    "total_contribucion": df_with_contr[contr_cols].sum().sort_values(ascending=False),
})
resumen_global["pct_de_ventas_predichas"] = (
    resumen_global["total_contribucion"] / df_with_contr["y_hat"].sum()
)

# Ejemplos de salida:
resumen_entidad.head()     # contribución total por canal en cada Entidad
resumen_global.head(10)    # top canales por contribución total (global)
betas_por_entidad.head()   # coeficientes (sensibilidad marginal) por Entidad
```
